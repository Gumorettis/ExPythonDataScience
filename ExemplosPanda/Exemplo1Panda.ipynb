```python
#importando bilbioteca do panda
import pandas as pd
```

```python
#importando um arquivo .csv para trabalhar carregando ela para o panda
dados = pd.read_csv("supermarket_sales.csv")
```

```python
dados #nome do objeto para chamar os dados
```

# Qual faturamento total por filial ? 

```python
#fazendo um agrupamento de dados por filial em cidades, ou seja, mostrar os dados de cada cidade separado
#depois Ã© pego a coluna 'total' para realizar a soma desses valores
dados.groupby("City")[["Total"]].sum()
```

```python
dados.groupby("City")[["Total","gross income"]].sum()
```

# Qual a percentual de participaÃ§Ã£o na receita de cada tipo de produto? 

```python
#Ã© aplicado uma soma de todos os produto 
(dados.groupby("Product line")[["Total"]].sum() / dados.groupby("Product line")[["Total"]].sum().sum()) * 100
```

# Como estÃ¡ distribuido o tipo de produto consumido por genero

```python
#Aqui serÃ¡ agrupado por nome do produto e genero, daÃ­ serÃ¡ somado a compra de cada genero de cada produto
#a funÃ§Ã£o pivot table serve para reorganizar as colunas da tabela gerada

dados.groupby(["Product line", "Gender"])[["Total"]].sum().pivot_table(index="Product line", columns="Gender")
```

# Qual foi o faturamento por mÃªs?

```python
#na linha abaixo estÃ¡ sendo feito uma formataÃ§Ã£o dos dados de data usando a funÃ§Ã£o datetime para poder trabalhar com mes
dados["Date"] = pd.to_datetime(dados["Date"]) #pega todas as datas existentes na tabela toda
#print(dados["Date"])

#criando abaixo novas colunas de mÃªs e ano, e organizando as datas em cada mÃªs usando apply e lambda para extrair sÃ³ o mÃªs
dados["Month"] = dados["Date"].apply(lambda x: x.month)

#print(dados["Month"])
```

```python
#agora faz o agrupamento por mÃªs e realiza a soma
dados.groupby(["Month"])[["Total"]].sum()
```

# Qual foi a mÃ©dia de avaliaÃ§Ã£o por cada filial em jan de 2019?

```python
#realizando um agrupamento especifico,
#acessando um dado especifico, no caso o ano e mÃªs dentro de uma data inteira, ano e mÃªs jÃ¡ foram criados anteriormente na tabela
dados[ (dados["Year"] == 2019 ) & (dados["Month"] == 1) ] ["Rating"].mean() #funÃ§Ã£o mean para realizar a mÃ©dia  
```

# Qual foi o gasto realizado por tipo de consumidor em cada filial? 

```python
# abaixo Ã© realizado um agrupamento duplo de consumidor e filial (branch)
dados.groupby( ["Customer type", "Branch"] ) [["Total"]].sum()
```

```python
#quanto custa em mÃ©dia os produtos por linha de produto (categoria)
dados.groupby("Product line")[["Unit price"]].mean()
```

```python
#Quantidade de produtos vendidos por categoria
dados.groupby("Product line")[["Quantity"]].sum()
```

```python
#media de gasto de cada tipo de consumidor em cada filial
dados.groupby(["Customer type", "Branch"])[["Total"]].mean()
```
